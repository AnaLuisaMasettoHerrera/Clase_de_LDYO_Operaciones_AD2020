Modelos de aprendizaje de máquina y promedios móviles.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
errores = pd.read_csv('metricas_finaleslab.csv')
```

```python
errores
```

```python
#Filtrar por entrenamiento y prueba
errores_ent= errores[errores['Conjunto']=='Entrenamiento']
errores_prueba= errores[errores['Conjunto']=='Prueba']
```

```python
errores_ent_mae= errores_ent[errores_ent['Metrica']== 'MAE']
```

```python
# Gráfica de errores de ENTRENAMIENTO 
plt.figure(figsize=(17,20))
plt.subplot(3, 1, 1)
plt.plot(errores_ent_mae.Mes, errores_ent_mae.MB_pm,'o-',color="red") #no aplica para entrenamiento
plt.plot(errores_ent_mae.Mes, errores_ent_mae.DT_1,'o-', color="forestgreen")
plt.plot(errores_ent_mae.Mes, errores_ent_mae.DT_5,'o-', color="orangered")
plt.plot(errores_ent_mae.Mes, errores_ent_mae.RF_100,'o-', color="cyan")
plt.plot(errores_ent_mae.Mes, errores_ent_mae.RF_500,'o-', color="blueviolet")

plt.legend(['MAE modelo base - Pedir el promedio de 2 meses','MAE Árbol de Decisión - 1 profundidad','MAE Árbol de Decisión - 5 profundidad','MAE Bosque aleatorio n=100','MAE Bosque aleatorio n=500'])
plt.ylim(0.05, 0.32)
plt.ylabel('MAE')
plt.title('Error Absoluto Medio de todos los Modelos')



plt.show()
```

```python
errores_prueba_mae= errores_prueba[errores_prueba['Metrica']=='MAE']
```

```python
# Gráfica de errores de PRUEBA 
plt.figure(figsize=(17,20))
plt.subplot(3, 1, 1)
plt.plot(errores_prueba_mae.Mes, errores_prueba_mae.MB_pm,'o-',color="crimson")
plt.plot(errores_prueba_mae.Mes, errores_prueba_mae.DT_1,'o-', color="forestgreen")
plt.plot(errores_prueba_mae.Mes, errores_prueba_mae.DT_5,'o-', color="orangered")
plt.plot(errores_prueba_mae.Mes, errores_prueba_mae.RF_100,'o-', color="cyan")
plt.plot(errores_prueba_mae.Mes, errores_prueba_mae.RF_500,'o-', color="blueviolet")

plt.legend(['MAE modelo base - Pedir el promedio de 2 meses','MAE Árbol de Decisión - 1 profundidad','MAE Árbol de Decisión - 5 profundidad','MAE Bosque aleatorio n=100','MAE Bosque aleatorio n=500'])
plt.ylim(0.07, 0.7)
plt.ylabel('MAE')
plt.title('Error Absoluto Medio de todos los Modelos')



plt.show()
```
