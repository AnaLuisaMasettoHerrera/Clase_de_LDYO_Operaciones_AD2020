```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np #determinar un rango de valores "clip"
from sklearn import tree #importar función para construir árboles
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error #errores que queremos calcular 
#from scipy.stats import sem
import warnings; warnings.simplefilter('ignore')#WARNING EN 3.3
from sklearn import metrics
```

```python
errores = pd.read_csv('E:/The Last Dance/Lab Diseño y Opt/resultados_finales_metricas_rf.csv')
```

```python
errores.head(10)
```

```python
#Filtrar por entrenamiento y prueba
errores_ent= errores[errores['Conjunto']=='entrenamiento']
errores_prueba= errores[errores['Conjunto']=='prueba']
```

```python
errores_ent_mae= errores_ent[errores_ent['Métrica']=='mae']

errores_prueba_mae= errores_prueba[errores_prueba['Métrica']=='mae']
```

```python
#Entrenamiento
plt.figure(figsize=(15,10))
plt.subplot(3, 1, 1)
plt.plot(errores_ent_mae.Mes, errores_ent_mae.Modelo_base,'o-',color="red") #no aplica para entrenamiento
plt.plot(errores_ent_mae.Mes, errores_ent_mae.rf_50,'o-', color="darkgreen")
plt.plot(errores_ent_mae.Mes, errores_ent_mae.rf_1_prof100,'o-', color="orange")
plt.legend(['MAE modelo base - Pedir el promedio de 2 meses','MAE Random Forest 50','MAE Random Forest 100'])
plt.ylim(0.001, 0.05)
plt.ylabel('Mae')
plt.title('Error Absoluto Medio de todos los Modelos')
```

```python
#Prueba
plt.figure(figsize=(15,10))
plt.subplot(3, 1, 1)
plt.plot(errores_prueba_mae.Mes, errores_prueba_mae.Modelo_base,'o-',color="red") #SI aplica para entrenamiento
plt.plot(errores_prueba_mae.Mes, errores_prueba_mae.rf_50,'o-', color="darkgreen")
plt.plot(errores_prueba_mae.Mes, errores_prueba_mae.rf_1_prof100,'o-', color="orange")
plt.legend(['MAE modelo base - Pedir el promedio de 2 meses','MAE Random Forest 50','MAE Random Forest 100'])
plt.ylim(0.001, 0.06)
plt.ylabel('Mae')
plt.title('Error Absoluto Medio de todos los Modelos')
```
